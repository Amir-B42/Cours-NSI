```python
def random_hide(p,k):
    if rd.random()<p:
        return ""
    else:
        return k
    
p = 0.5
print(random_hide(p,3))
    
```

On comprends que random prends un nombre aléatoire entre 0 et 1, et que cette fonction va aléatoirement, soit ne rien renvoyer(c'est comme si on "supprimait" k), soit renvoyer k. On peut décider de la probabilité en affectant un nombre plus ou moins grand à p. 

### 9) Faire une fonction qui prend en entrée un tableau qui affiche valeur par valeur toutes les valeurs du tableau. (différent de AfficherTableau qui affiche les valeurs 3 par 3)

```python
def Valeurs(M):
    nb = []
    for i in range(len(M)):
        for j in range(len(M[i])):
            nb.append(M[i][j])
    return nb


Valeurs([[1,2,3],[4,5,6],[7,8,9]])
```
