# Exercice 3

Le tableau suivant donne la mention en fonction de la note N obtenue à la première session du baccalauréat.

|Condition | Mention|
|---|---|
|$0\leq N <8$ | Ajournée |
|$8 \leq N <10$ | Rattrapage |
|$10 \leq N < 12$ | Sans Mention |
|$12 \leq N < 14$ | Assez Bien |
|$14 \leq N < 16$ | Bien|
|$N \geq 16$ |Très bien |

Définissez une fonction Python qui prend une note en paramètre et renvoie la men-
tion correspondante sous forme de chaîne de caractère.

````python
def mention_bac(moy): 
    if moy < 8:
        print("Ajournée")
    elif 8 <= moy < 10:
        return "Vous êtes en rattrapage"
    elif 10 <= moy <12 :
        return "Vous avez eu votre bac sans mention"
    elif 12 <= moy <14 :
        return "Vous avez eu votre bac avec mention assez bien"
    elif 14 <= moy <16 :
        return "Vous avez eu botre bac avec mention bien"
    elif moy >= 16:
        return "Vous avez eu votre bac avec mention très bien"
mention_bac(16)
````
